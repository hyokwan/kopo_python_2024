```python
loadedModel = YOLO(model="yolov8n.pt")
framePath = "./yolotestset_20240522/vita500_man.png"
```

```python
frame = cv2.imread(  framePath, cv2.IMREAD_COLOR )
detection = loadedModel( frame )[0]
boxInfos = detection.boxes.data.tolist()
for i in range(0, len(boxInfos)):
    # 디버깅 포인트 반복문을 돌리기 위해 내부 코드 까보기
    data = boxInfos[i]
    # 좌표체계 확인
    xone = int(data[0])
    yone = int(data[1])
    xtwo = int(data[2])
    ytwo = int(data[3])
    cv2.rectangle( frame,
                   (xone, yone),
                   (xtwo, ytwo),
                   (0,255,0),
                   2)
    # 클래스 가져오기
    ## data [xone, yone, xtwo, ytwo, confidence, class, ]
    classIndex = int( data[5] )
    className = loadedModel.names
    textValue = className[ classIndex ]
    textValue
    ### confidence_scre
    confidenceScore = str( round( data[4], 2) )
    confidenceScore
    putTextValue = textValue+"_"+confidenceScore
    putTextValue
    cv2.putText(frame,
                putTextValue,
                (xone, yone),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (255,255,255),
                2)
```

#### 이미지 영상에 어떻게 적용하지

```python
# 영상 파일 경로
video_path = './yolotestset_20240522/testVideo.mp4'
# 저장할 동영상 파일 경로 및 설정
output_path = './yolotestset_20240522/output_video_yolo.mp4'
fps = 30  # 저장할 동영상의 프레임 속도 (FPS)
# 비디오 캡처 객체 생성
capture = cv2.VideoCapture(video_path)
# 프레임 크기 가져오기
frame_width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 비디오 라이터 객체 생성
fcc = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter(output_path, fcc, fps, (frame_width,frame_height))
# 프레임을 하나씩 읽어와서 저장
while capture.isOpened():
    ret, frame = capture.read()
    if ret:

        # frame가공후 frame을 리턴하자!!!
        # frame = cv2.imread(  framePath, cv2.IMREAD_COLOR )
        detection = loadedModel( frame )[0]
        boxInfos = detection.boxes.data.tolist()
        for i in range(0, len(boxInfos)):
            # 디버깅 포인트 반복문을 돌리기 위해 내부 코드 까보기
            data = boxInfos[i]
            # 좌표체계 확인
            xone = int(data[0])
            yone = int(data[1])
            xtwo = int(data[2])
            ytwo = int(data[3])
            cv2.rectangle( frame,
                        (xone, yone),
                        (xtwo, ytwo),
                        (0,255,0),
                        2)
            # 클래스 가져오기
            ## data [xone, yone, xtwo, ytwo, confidence, class, ]
            classIndex = int( data[5] )
            className = loadedModel.names
            textValue = className[ classIndex ]
            textValue
            ### confidence_scre
            confidenceScore = str( round( data[4], 2) )
            confidenceScore
            putTextValue = textValue+"_"+confidenceScore
            putTextValue
            cv2.putText(frame,
                        putTextValue,
                        (xone, yone),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (255,255,255),
                        2)
        # 프레임 저장
        video.write(frame)
    else:
        break

# 비디오 캡처 및 라이터 객체 해제
capture.release()
video.release()
```
